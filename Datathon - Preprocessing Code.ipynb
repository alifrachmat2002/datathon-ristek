# Process and Prepare the Data to Suit the Need of the Model
### Mengolah dan Menyiapkan Data Sesuai Kebutuhan Model

 

```python
#Tambah kolom rute, yakni gabungan dari titik awal dan titik akhir
#Add a 'rute'(route) column, which is a combination of the starting point and ending point


train_data_w_route = train_data
train_data_w_route['rute'] = train_data_w_route['id_titik_mulai'].astype(str) + '-' + train_data_w_route['id_titik_akhir'].astype(str)
jml_rute_unik = len(train_data_w_route['rute'].unique())

print(f'jumlah rute unik yang ada di dalam data training adalah {jml_rute_unik}')
train_data_w_route.info()
```

```python
#Tambah kolom rute, yakni gabungan dari titik awal dan titik akhir

# test_data_w_route = test_data
# test_data_w_route['rute'] = test_data_w_route['id_titik_mulai'].astype(str) + '-' + test_data_w_route['id_titik_akhir'].astype(str)
# jml_rute_unik = len(test_data_w_route['rute'].unique())

print(f'jumlah rute unik yang ada di dalam data testing adalah {jml_rute_unik}')
# the number of unique routes in the testing data is 934
```

```python
#Ubah tipe data pada kolom id_jalan, id_titik_mulai dan id_titik_akhir untuk menghindari agregasi
#Change the data type in the id_jalan, id_titik_mulai and id_titik_akhir columns to avoid aggregation

train_data_w_route['id_jalan'] = train_data_w_route['id_jalan'].astype(str) 
train_data_w_route['id_titik_mulai'] = train_data_w_route['id_titik_mulai'].astype(str) 
train_data_w_route['id_titik_akhir'] = train_data_w_route['id_titik_akhir'].astype(str)
print(train_data_w_route.info())

test_data_w_route['id_jalan'] = test_data_w_route['id_jalan'].astype(str) 
test_data_w_route['id_titik_mulai'] = test_data_w_route['id_titik_mulai'].astype(str) 
test_data_w_route['id_titik_akhir'] = test_data_w_route['id_titik_akhir'].astype(str)
print(test_data_w_route.info())
```

```python
routes = pd.unique(train_data_w_route['rute'])
routes[:10]
```

```python
#mengelompokkan data berdasarkan rutenya dalam list, 
#tiap elemen list berisi 1 tabel khusus 1 rute 

# grouping data based on its route in the list,
#each list element contains 1 custom table for each route


#train_data
train_data_grouped = []
for route in routes:
     train_data_grouped.append(train_data_w_route[train_data_w_route['rute'] == route])


#contoh
print(train_data_grouped[0],"\n")

#test data
test_data_grouped = []
for route in routes:
     test_data_grouped.append(test_data_w_route[test_data_w_route['rute'] == route])


print(test_data_grouped[0],"\n")
```

```python
# Ternyata ada data yang tidak lengkap pada train_data dan test_data. 
# Semestinya data diambil dan direkam tiap 1 jam, 
# nyatanya ada beberapa "jam" yang tidak ada di dalam data 

# It turns out that there is incomplete data in train_data and test_data.
# Data should be taken and recorded every 1 hour,
# in fact there are some "hour" that are not in the data
```

```python
#Melengkapi data yang tidak ada
# Completing missing data

# membuat list timestamps untuk mengisi data tanggal-jam yang kosong
# create a list of timestamps to fill in the empty date-time data

start_date = pd.Timestamp('2020-02-01 00:00:00+00:00')
end_date = pd.Timestamp('2020-02-22 23:00:00+00:00')
time_interval = pd.DateOffset(hours=1)
timestamps = pd.date_range(start=start_date, end=end_date, freq=time_interval)

# mengubah timestamps yang sudah dibuat menjadi DataFrame
# change the created timestamps to DataFrame
generated_data = pd.DataFrame({'waktu_setempat': timestamps})

#ini data timestamp yang sudah dibuat
#this is the timestamp data that has been created
generated_data
```

```python
#mencoba mengisi data yang kosong dengan salah satu data train, 
#tanpa mengubah data train-nya terlebih dulu

#a test to try to fill in the blank data with one of the training data, without changing the train data first

test_merge_data = train_data_grouped[0]

merged_data = generated_data.merge(test_merge_data, on='waktu_setempat', how='left')

merged_data
```

```python
#mengisi data kosong menggunakan ffill dan bfill
#fill empty data using ffill and bfill

merged_data['rerata_kecepatan'] = merged_data['rerata_kecepatan'].fillna(method='ffill')
merged_data['rerata_kecepatan'] = merged_data['rerata_kecepatan'].fillna(method='bfill')

merged_data['id_jalan'] = merged_data['id_jalan'].fillna(method='ffill')
merged_data['id_jalan'] = merged_data['id_jalan'].fillna(method='bfill')

merged_data['id_titik_mulai'] = merged_data['id_titik_mulai'].fillna(method='ffill')
merged_data['id_titik_mulai'] = merged_data['id_titik_mulai'].fillna(method='bfill')

merged_data['id_titik_akhir'] = merged_data['id_titik_akhir'].fillna(method='ffill')
merged_data['id_titik_akhir'] = merged_data['id_titik_akhir'].fillna(method='bfill')

merged_data
```

```python
#dan sudah tidak ada data yang kosong pada tabel, 
#horeeeeee

#no more empty data in the table,
#hooray

merged_data.isna().sum()
```

```python
#melakukan prosedur pengisian data pada train_data dan test_data

#train_data
sum = 0
train_data_filled = []
for data in train_data_grouped:
    
    #rute = data['rute'].unique()
    merged_data = generated_data.merge(data, on='waktu_setempat', how='left')
    
    merged_data['rerata_kecepatan'] = merged_data['rerata_kecepatan'].fillna(method='ffill')
    merged_data['rerata_kecepatan'] = merged_data['rerata_kecepatan'].fillna(method='bfill')

    merged_data['id_jalan'] = merged_data['id_jalan'].fillna(method='ffill')
    merged_data['id_jalan'] = merged_data['id_jalan'].fillna(method='bfill')

    merged_data['id_titik_mulai'] = merged_data['id_titik_mulai'].fillna(method='ffill')
    merged_data['id_titik_mulai'] = merged_data['id_titik_mulai'].fillna(method='bfill')
    
    merged_data['id_titik_akhir'] = merged_data['id_titik_akhir'].fillna(method='ffill')
    merged_data['id_titik_akhir'] = merged_data['id_titik_akhir'].fillna(method='bfill')
    
    merged_data['rute'] = merged_data['rute'].fillna(method='ffill')
    merged_data['rute'] = merged_data['rute'].fillna(method='bfill')
    
    #print(f'data yang kosong untuk rute {rute} sudah diisi')
    train_data_filled.append(merged_data)
    sum += 1
    
print(f'ada {sum} tabel yang sudah diisi')
```

```python
#Cek salah satu data
#Check one of the data

train_data_filled[933]

#everything looks ok
```

```python
#gabungkan semua data menjadi 1 tabel DataFrame
#combine all data into 1 DataFrame table

train_data_ready = pd.concat(train_data_filled, axis=0, ignore_index=True)
train_data_ready
```

```python
#export data yang sudah di preprocess menjadi file csv
#export preprocessed data into a csv file 
train_data_ready.to_csv('output_file.csv', index=False)
```
